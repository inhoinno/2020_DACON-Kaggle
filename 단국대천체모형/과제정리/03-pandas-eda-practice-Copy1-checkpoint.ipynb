## 시험데이터 로드

```python
tst = pd.read_csv(tst_file, index_col=0)
print(tst.shape)
tst.head()
```

## 학습/시험 데이터 결합

```python
df = pd.concat([trn, tst], axis=0)
print(df.shape)
df.tail()
```

```python
df.fillna(-1, inplace=True)
df.tail()
```

## 피쳐변환

```python
df['nObserve'].hist(bins=30)
```

```python
df['nObserve'] = df['nObserve'].apply(np.log1p)
```

```python
df['nObserve'].hist(bins=30)
```

## 피쳐 생성

```python
# 빼기 시작
df['d_dered_u'] = df['dered_u'] - df['u']
df['d_dered_g'] = df['dered_g'] - df['g']
df['d_dered_r'] = df['dered_r'] - df['r']
df['d_dered_i'] = df['dered_i'] - df['i']
df['d_dered_z'] = df['dered_z'] - df['z']
df['d_dered_rg'] = df['dered_r'] - df['dered_g']
df['d_dered_ig'] = df['dered_i'] - df['dered_g']
df['d_dered_zg'] = df['dered_z'] - df['dered_g']
df['d_dered_ri'] = df['dered_r'] - df['dered_i']
df['d_dered_rz'] = df['dered_r'] - df['dered_z']
df['d_dered_iz'] = df['dered_i'] - df['dered_z']
df['d_obs_det'] = df['nObserve'] - df['nDetect']
# 더하기 시작
df['p_dered_u'] = df['dered_u'] + df['u']
df['p_dered_g'] = df['dered_g'] + df['g']
df['p_dered_r'] = df['dered_r'] + df['r']
df['p_dered_i'] = df['dered_i'] + df['i']
df['p_dered_z'] = df['dered_z'] + df['z']
df['p_dered_rg'] = df['dered_r'] + df['dered_g']
df['p_dered_ig'] = df['dered_i'] + df['dered_g']
df['p_dered_zg'] = df['dered_z'] + df['dered_g']
df['p_dered_ri'] = df['dered_r'] + df['dered_i']
df['p_dered_rz'] = df['dered_r'] + df['dered_z']
df['p_dered_iz'] = df['dered_i'] + df['dered_z']
df['p_obs_det'] = df['nObserve'] + df['nDetect']
# 나누기 시작
df['n_dered_u'] = df['dered_u'] / df['u']
df['n_dered_g'] = df['dered_g'] / df['g']
df['n_dered_r'] = df['dered_r'] / df['r']
df['n_dered_i'] = df['dered_i'] / df['i']
df['n_dered_z'] = df['dered_z'] / df['z']
df['n_dered_rg'] = df['dered_r'] / df['dered_g']
df['n_dered_ig'] = df['dered_i'] / df['dered_g']
df['n_dered_zg'] = df['dered_z'] / df['dered_g']
df['n_dered_ri'] = df['dered_r'] / df['dered_i']
df['n_dered_rz'] = df['dered_r'] / df['dered_z']
df['n_dered_iz'] = df['dered_i'] / df['dered_z']
df['n_obs_det'] = df['nObserve'] / df['nDetect']

print(df.shape)
df.head()
```

## 피쳐 삭제

```python
df.corr().style.background_gradient()
```

```python
df.drop(['airmass_z', 'airmass_i', 'airmass_r', 'airmass_g', 'u', 'g', 'r', 'i', 'nDetect', 'd_dered_rg', 'd_dered_ri'], 
        axis=1, inplace=True)
print(df.shape)
df.head()
```
